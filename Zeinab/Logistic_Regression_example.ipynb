```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x = np.array([2.4,5.0,1.5,3.8,8.7,3.6,1.2,8.1,2.5,5,1.6,1.6,2.4,3.9,5.4])
y = np.array([2.1,4.7,1.7,3.6,8.7,3.2,1,8,2.4,6,1.1,1.3,2.4,3.9,4.8])
n = np.size(x)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x,y)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
a0 = 0  # intercept
a1 = 0  # slope
lr = 0.001  # learning rate
iterations = 1000  # number of iterations
error = []

for itr in range(iterations):
    error_cost = 0
    cost_a0 = 0
    cost_a1 = 0
    for i in range(len(x)):
        y_pred = a0 + a1 * x[i]
        error_cost = error_cost + (y[i] - y_pred) ** 2
        partial_wrt_a0 = -2 * (y[i] - (a0 + a1 * x[i]))
        partial_wrt_a1 = -2 * x[i] * (y[i] - (a0 + a1 * x[i]))
        cost_a0 = cost_a0 + partial_wrt_a0
        cost_a1 = cost_a1 + partial_wrt_a1
    a0 = a0 - lr * cost_a0
    a1 = a1 - lr * cost_a1
    error.append(error_cost)

    # Print the progress at every 100 iterations
    if itr % 100 == 0:
        print(f"Iteration {itr}: a0 = {a0}, a1 = {a1}, Error = {error_cost}")
```

```python
print("Final coefficients:")
print(f"a0 = {a0}, a1 = {a1}")
```

```python
plt.plot(range(1, iterations + 1), error, color='r')
plt.xlabel("Iterations")
plt.ylabel("Error")
plt.title("Gradient Descent for Linear Regression")
plt.show()
```

```python
pred = a0 + a1*x
print(pred)
```

```python
plt.scatter(x,y, color='r')
plt.plot(x, pred, color='g')
plt.show()
```

```python
error1 = y - pred
se = np.sum(error1**2)
mse = se/n
print(mse)
print(a0)
print(a1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
x = x.reshape(-1,1)
model = LinearRegression()
model.fit(x,y)
salary_pred = model.predict(x)
mse = mean_squared_error(y, salary_pred)
print('slope', model.coef_)
print('intercept', model.intercept_)
print('MSE', mse)
```
