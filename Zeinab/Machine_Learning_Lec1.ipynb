# Data Processing - Diabetes Dataset

## Step 1: Importing the packages for data processing and loading the data

```python
#Step 1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("diabetes.csv")
data.head()
```

## Step 2: Data Overview
### We begin by loading the dataset and displaying basic information like the number of rows, columns, data types, and non-null counts. This helps us understand the dataset's structure.
### We calculate and display summary statistics for numerical columns, which include measures like mean, median, and quartiles. This provides an overview of the central tendency and spread of the data.

```python
# Step 2: Data Overview
# Display basic information about the dataset
print("Step 2: Data Overview")
print(data.info())
```

# Step 3: Summary Statistics
### We calculate and display summary statistics for numerical columns, which include measures like mean, median, and quartiles. This provides an overview of the central tendency and spread of the data.

```python
# Step 3: Summary Statistics
# Display summary statistics (mean, median, min, max, etc.) for numerical columns
print("\nStep 3: Summary Statistics")
print(data.describe())
```

# Step 4: Missing Values
###  We check for missing values in the dataset and display the count of missing values for each column. Identifying missing data is crucial for data preprocessing.

```python
# Step 4: Missing Values
# Check for missing values and visualize them if any
print("\nStep 4: Missing Values")
print(data.isnull().sum())
```

# Step 5: Data Distribution
### We visualize the distribution of numerical features using histograms and kernel density plots. This helps us understand the data's shape and potential outliers.

```python
# Step 5: Data Distribution
# Visualize the distribution of numerical features
print("\nStep 5: Data Distribution")

# Get the list of numerical columns
numerical_columns = data.select_dtypes(include=['int', 'float']).columns
num_numerical_columns = len(numerical_columns)

# Determine the subplot layout based on the number of numerical columns
num_rows = (num_numerical_columns - 1) // 3 + 1
num_cols = min(num_numerical_columns, 3)

# Create subplots dynamically
plt.figure(figsize=(15, 5 * num_rows))
for i, column in enumerate(numerical_columns):
    plt.subplot(num_rows, num_cols, i + 1)
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

# Step 6: Correlation Analysis
### We calculate and visualize correlations between numerical features using a heatmap. This step helps identify relationships between variables.

```python
# Step 6: Correlation Analysis
# Calculate and visualize correlations between numerical features
print("\nStep 6: Correlation Analysis")
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# Step 7: Outliers
### We use box plots to detect and visualize outliers in numerical columns. Outliers can significantly impact modeling and need to be addressed.

```python
# Step 7: Outliers
# Detect and visualize outliers using box plots
print("\nStep 7: Outliers")

# Get the list of numerical columns
numerical_columns = data.select_dtypes(include=['int', 'float']).columns
num_numerical_columns = len(numerical_columns)

# Determine the subplot layout based on the number of numerical columns
num_rows = (num_numerical_columns - 1) // 3 + 1
num_cols = min(num_numerical_columns, 3)

# Create subplots dynamically
plt.figure(figsize=(15, 5 * num_rows))
for i, column in enumerate(numerical_columns):
    plt.subplot(num_rows, num_cols, i + 1)
    sns.boxplot(data=data, x=column)
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```

# Step 8: Pairwise Relations
### We create a pairplot to visualize pairwise relationships between numerical features. This can reveal patterns or trends in the data.

```python
# Step 8: Pairwise Relationships
# Visualize pairwise relationships between numerical features
print("\nStep 8: Pairwise Relationships")
sns.pairplot(data=data, hue='Outcome', diag_kind='kde')
plt.suptitle("Pairwise Relationships")
plt.show()
```

# Step 9: Target Counts
### We count the total number of Target variables. This is done especially for the case of binary and multiclass classification

```python
# Step 9: Target Counts
# Count the number of occurrences of 0 and 1 in the "outcome" column
print("\nStep 9: Pairwise Relationships")
outcome_counts = data['Outcome'].value_counts()

# Create a bar plot
plt.figure(figsize=(8, 6))
outcome_counts.plot(kind='bar', color=['blue', 'green'])
plt.title('Count of 0 and 1 in Outcome')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Remove rotation of x-axis labels
plt.show()
```
